```python
import seaborn as sns
```

```python
# Basic plotting
df = sns.load_dataset('tips')
df
```

```python
# Scatter plot

import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',
                y='tip',
                data=df)
plt.title("Scatter plot")
plt.show()
```

```python
# Line plot

sns.lineplot(x='size',
             y='total_bill',
             data=df)
plt.show()
```

```python
# Categorical plots

sns.barplot(x='day',
            y='total_bill',
            data=df)
plt.show()
```

```python
# Box plot

sns.boxplot(x='day',
            y='total_bill',
            data=df)
plt.show()
```

```python
# Violin plot

sns.violinplot(x='day',
               y='total_bill',
               data=df)
plt.show()
```

```python
# Histogram

sns.histplot(df['total_bill'],
             bins=10,
             kde=True)
plt.show()
```

```python
# Pairplot

sns.pairplot(df)
plt.show()
```

```python
# heatmap

corr = df[['total_bill','tip','size']].corr()
sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')
plt.show()
```
