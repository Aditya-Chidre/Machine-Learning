```python
import numpy as np
marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
```

```python
minimum,q1,medinan,q3,maximum =np.quantile(marks,[0,0.25,0.50,0.75,1])
minimum,q1,medinan,q3,maximum
```

```python
iqr = q3-q1
iqr
```

```python
lower_fence=q1-1.5*(iqr)
higher_fence=q3+1.5*(iqr)
lower_fence, higher_fence
```
